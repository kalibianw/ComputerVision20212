```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os

IMG_PATH_1 = "../resource/img/homework 4/dsu4.jpg"
print(os.path.exists(IMG_PATH_1))
IMG_PATH_2 = "../resource/img/homework 4/dsu1.jpg"
print(os.path.exists(IMG_PATH_2))
```

```python
img = cv2.imread(IMG_PATH_2)
rows, cols, ch = img.shape

pts1 = np.float32([[237, 269], [570, 235], [566, 496], [235, 462]])
pts2 = np.float32([[200, 200], [500, 200], [500, 400], [200, 400]])

M = cv2.getPerspectiveTransform(pts1, pts2)

dst = cv2.warpPerspective(img, M, (1024, 768))

plt.subplot(121)
plt.imshow(img)
plt.title('Input')

plt.subplot(122)
plt.imshow(dst)
plt.title('Output')
plt.show()
```
