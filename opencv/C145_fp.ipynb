```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

IMG_PATH = "../resource/img/fingerprint/1128-400x300.jpg"
if os.path.exists(IMG_PATH):
    print("True")
else:
    print("False")
img = cv2.imread(IMG_PATH, 0)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 2))

gen = cv2.erode(img, kernel, iterations=1)

plt.subplot(121)
plt.imshow(img)
plt.title("Original")
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(gen)
plt.title("Generating")
plt.xticks([])
plt.yticks([])

plt.show()
```
