```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os

IMG_PATH = "../resource/img/homework 4/dsu1.jpg"
print(os.path.exists(IMG_PATH))
```

```python
img = cv2.imread(IMG_PATH)
rows, cols, ch = img.shape

pts1 = np.float32([[505, 324], [763, 320], [326, 529], [759, 505]])
pts2 = np.float32([[450, 330], [750, 300], [300, 500], [730, 500]])

M = cv2.getPerspectiveTransform(pts1, pts2)

dst = cv2.warpPerspective(img, M, (1024, 768))

plt.subplot(131)
plt.imshow(img)
plt.title('Input')

plt.subplot(132)
plt.imshow(dst)
plt.title('Output')
plt.show()

plt.subplot(133)
plt.imshow(dst[350:500, 270:750])
plt.title('Output')
plt.show()
```
